```python
import pandas as pd

# Load the DataFrame
file_path = r'C:\Users\USER\Desktop\data science\Coordinated project\cleaned_crimes_in_la_with_avg_age.csv'
```

```python
df = pd.read_csv(file_path, dtype=str)
```

```python
df
```

```python
df.columns
```

```python
df['Vict Age'] = pd.to_numeric(df['Vict Age'], errors='coerce')
```

```python
df = df[df['Vict Age'] > 0]

# Optionally reset the index if needed
df.reset_index(drop=True, inplace=True)
```

```python

# Display the updated dataframe
print(df)
```

```python
bins = [0, 13, 20, 36, 61, float('inf')]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
```

```python
df.loc[:, 'Age Group'] = pd.cut(df['Vict Age'], bins=bins, labels=labels, right=False)
```

```python
print(df[['Vict Age', 'Age Group']])
```

```python
df
```

```python
cleaned_file_path = r'C:\Users\USER\Desktop\data science\Coordinated project\UPDATED-cleaned_crimes_in_la_with_avg_age.csv'
try:
    df.to_csv(cleaned_file_path, index=False)
    print("Data cleaned and saved successfully!")
except Exception as e:
    print(f"Error saving cleaned data: {e}")
```
