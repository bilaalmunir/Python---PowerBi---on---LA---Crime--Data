```python
import pandas as pd

# Load the DataFrame
file_path = r'C:\Users\USER\Desktop\data science\Coordinated project\Crime_Data_from_2020_to_Present_20240630.csv'
```

```python
try:
    df = pd.read_csv(file_path, dtype=str, low_memory=False)
    print(f"DataFrame loaded successfully:\n{df.head()}")
except FileNotFoundError:
    print(f"File not found at {file_path}")
except Exception as e:
    print(f"Error loading the DataFrame: {e}")
    raise

df['Vict Age'] = pd.to_numeric(df['Vict Age'], errors='coerce')
```

```python
invalid_age_rows = df[df['Vict Age'] <= 0]
print(f"Rows with invalid 'Vict Age':\n{invalid_age_rows}")
```

```python
for idx, row in invalid_age_rows.iterrows():
    crime_desc = row['Crm Cd Desc']
    try:
        # Calculate the average Vict Age for the same Crime Cd Desc with valid ages
        valid_ages = df[(df['Crm Cd Desc'] == crime_desc) & (df['Vict Age'] > 0)]['Vict Age']
        if not valid_ages.empty:
            avg_age = valid_ages.mean()
            df.at[idx, 'Vict Age'] = avg_age
            print(f"Row {idx} updated with average age {avg_age}")
        else:
            print(f"No valid ages found for Crime Cd Desc {crime_desc} in row {idx}")
    except Exception as e:
        print(f"Error processing row {idx}: {e}")
```

```python
cleaned_file_path = r'C:\Users\USER\Desktop\data science\Coordinated project\cleaned_crimes_in_la_with_avg_age.csv'
try:
    df.to_csv(cleaned_file_path, index=False)
    print("Data cleaned and saved successfully!")
except Exception as e:
    print(f"Error saving cleaned data: {e}")
```
